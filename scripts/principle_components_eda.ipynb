```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# =========================
# 1) DOSYA YOLLARI
# =========================

DATA_FILES = {
    "Fillmore": ".\\selected_speeches\\fillmore_core5_all_blocks_emfd_vectors.csv",
    "Ford": ".\\selected_speeches\\ford_core5_all_blocks_emfd_vectors.csv",
    "Truman": ".\\selected_speeches\\truman_core5_all_blocks_emfd_vectors.csv",
    "Eisenhower": ".\\selected_speeches\\eisenhower_core5_all_blocks_emfd_vectors.csv",
}

PCA_FILE = ".\\pca_export\\pca_export_summary.xlsx"

OUT_DIR = Path(".\pc_projection_plots")
OUT_DIR.mkdir(exist_ok=True)

FEATURES = ["care_p", "fairness_p", "loyalty_p", "authority_p", "sanctity_p"]
PCS = ["PC1", "PC2", "PC3", "PC4"]


# =========================
# 2) PCA LOADINGS VE SCALER OKU
# =========================

loadings_df = pd.read_excel(PCA_FILE, sheet_name="loading_matrix_aligned")
scaler_df = pd.read_excel(PCA_FILE, sheet_name="scaler_stats")

loadings_df = loadings_df.set_index("feature").loc[FEATURES]
scaler_df = scaler_df.set_index("feature").loc[FEATURES]

means = scaler_df["mean"].values.astype(float)
scales = scaler_df["scale"].values.astype(float)

pc_vectors = {
    pc: loadings_df[pc].values.astype(float)
    for pc in PCS
}


# =========================
# 3) PC PROJECTION / SCORE HESABI
# =========================

all_dfs = []

for president, path in DATA_FILES.items():
    df = pd.read_csv(path)

    X_raw = df[FEATURES].values.astype(float)

    # PCA standartlaştırılmış veriyle kurulduğu için aynı dönüşüm uygulanır.
    X_z = (X_raw - means) / scales

    # PCA eksenleri unit vector olduğu için projection = dot product
    for pc, v in pc_vectors.items():
        df[f"score_{pc}"] = X_z @ v

    df["president_short"] = president
    all_dfs.append(df)

df_all = pd.concat(all_dfs, ignore_index=True)

out_csv = OUT_DIR / "four_presidents_PC1_PC4_projection_scores.csv"
df_all.to_csv(out_csv, index=False)

print("Projection score dosyası kaydedildi:")
print(out_csv)


# =========================
# 4) HER KONUŞMA İÇİN PC1-PC4 TEK GRAFİK
# =========================

for president, df_p in df_all.groupby("president_short"):

    for speech, df_s in df_p.groupby("speech_label"):
        df_s = df_s.sort_values("block_idx")

        plt.figure(figsize=(14, 6))

        for pc in PCS:
            plt.plot(
                df_s["block_idx"],
                df_s[f"score_{pc}"],
                linewidth=1.8,
                label=pc
            )

        plt.axhline(0, linewidth=1, linestyle="--")
        plt.title(f"{president} - {speech} | PC1-PC4 Projection Scores")
        plt.xlabel("Block index")
        plt.ylabel("PCA projection score")
        plt.legend()
        plt.tight_layout()

        safe_speech = (
            str(speech)
            .replace("/", "-")
            .replace("\\", "-")
            .replace(":", "-")
            .replace("*", "-")
            .replace("?", "")
            .replace('"', "")
            .replace("<", "")
            .replace(">", "")
            .replace("|", "")
        )

        out_png = OUT_DIR / f"{president}_{safe_speech}_PC1_PC4_projection.png"
        plt.savefig(out_png, dpi=300)
        plt.close()


# =========================
# 5) HER BAŞKAN İÇİN HER PC'Yİ 5 KONUŞMADA KARŞILAŞTIR
# =========================

for president, df_p in df_all.groupby("president_short"):

    for pc in PCS:
        plt.figure(figsize=(14, 6))

        for speech, df_s in df_p.groupby("speech_label"):
            df_s = df_s.sort_values("block_idx")

            plt.plot(
                df_s["block_idx"],
                df_s[f"score_{pc}"],
                linewidth=1.5,
                label=speech
            )

        plt.axhline(0, linewidth=1, linestyle="--")
        plt.title(f"{president} | {pc} Projection Scores Across Speeches")
        plt.xlabel("Block index")
        plt.ylabel(f"{pc} projection score")
        plt.legend(fontsize=8)
        plt.tight_layout()

        out_png = OUT_DIR / f"{president}_{pc}_all_speeches_projection.png"
        plt.savefig(out_png, dpi=300)
        plt.close()


# =========================
# 6) KONUŞMA BAZLI ÖZET TABLO
# =========================

summary = (
    df_all
    .groupby(["president_short", "speech_label"])[[f"score_{pc}" for pc in PCS]]
    .agg(["mean", "median", "std", "min", "max"])
)

summary_csv = OUT_DIR / "PC1_PC4_projection_summary_by_speech.csv"
summary.to_csv(summary_csv)

print("Plotlar ve özet tablo kaydedildi:")
print(OUT_DIR)
print(summary_csv)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pathlib import Path

# =========================
# 1) DOSYA YOLLARI
# =========================

DATA_FILES = {
    "Fillmore": ".\\selected_speeches\\fillmore_core5_all_blocks_emfd_vectors.csv",
    "Ford": ".\\selected_speeches\\ford_core5_all_blocks_emfd_vectors.csv",
    "Truman": ".\\selected_speeches\\truman_core5_all_blocks_emfd_vectors.csv",
    "Eisenhower": ".\\selected_speeches\\eisenhower_core5_all_blocks_emfd_vectors.csv",
}

PCA_FILE = ".\\pca_export\\pca_export_summary.xlsx"

OUT_DIR = Path(".\\pc_gpy_projection_plotly")
OUT_DIR.mkdir(exist_ok=True)

FEATURES = ["care_p", "fairness_p", "loyalty_p", "authority_p", "sanctity_p"]
PCS = ["PC1", "PC2", "PC3", "PC4", "PC5"]
AXES = PCS + ["GPY"]


# =========================
# 2) PCA LOADINGS VE SCALER OKU
# =========================

loadings_df = pd.read_excel(PCA_FILE, sheet_name="loading_matrix_aligned")
scaler_df = pd.read_excel(PCA_FILE, sheet_name="scaler_stats")

loadings_df = loadings_df.set_index("feature").loc[FEATURES]
scaler_df = scaler_df.set_index("feature").loc[FEATURES]

means = scaler_df["mean"].values.astype(float)
scales = scaler_df["scale"].values.astype(float)

pc_vectors = {
    pc: loadings_df[pc].values.astype(float)
    for pc in PCS
}


# =========================
# 3) GPY EKSENİ
# =========================

gonye = np.array([0.808, 0.539, 0.153, 0.179, 0.000], dtype=float)
pergel = np.array([0.103, 0.767, 0.169, 0.614, 0.153], dtype=float)
yasa = np.array([0.080, 0.000, 0.577, 0.742, 0.330], dtype=float)

gpy_raw = gonye + pergel + yasa
gpy_norm = gpy_raw / np.linalg.norm(gpy_raw)

print("GPY raw:", gpy_raw)
print("GPY norm:", np.linalg.norm(gpy_raw))
print("GPY normalized:", gpy_norm)
print("GPY normalized norm:", np.linalg.norm(gpy_norm))


# =========================
# 4) PC + GPY PROJECTION / SCORE HESABI
# =========================

all_dfs = []

for president, path in DATA_FILES.items():
    df = pd.read_csv(path)

    X_raw = df[FEATURES].values.astype(float)

    # PCA skorları için standartlaştırılmış veri gerekir.
    X_z = (X_raw - means) / scales

    # PC projection = standardized block vector dot PCA axis
    for pc, v in pc_vectors.items():
        df[f"score_{pc}"] = X_z @ v

    # GPY projection = raw moral vector dot normalized GPY axis
    # GPY dışsal semantik eksen olduğu için ham eMFD uzayında hesaplanır.
    df["score_GPY"] = X_raw @ gpy_norm

    df["president_short"] = president
    all_dfs.append(df)

df_all = pd.concat(all_dfs, ignore_index=True)

out_csv = OUT_DIR / "four_presidents_PC1_PC5_GPY_projection_scores.csv"
df_all.to_csv(out_csv, index=False)

print("Projection score dosyası kaydedildi:")
print(out_csv)


# =========================
# 5) GÜVENLİ DOSYA ADI
# =========================

def safe_filename(x):
    return (
        str(x)
        .replace("/", "-")
        .replace("\\", "-")
        .replace(":", "-")
        .replace("*", "-")
        .replace("?", "")
        .replace('"', "")
        .replace("<", "")
        .replace(">", "")
        .replace("|", "")
        .replace("\n", " ")
        .strip()
    )


# =========================
# 6) HER KONUŞMA İÇİN PC1-PC5 + GPY TEK HTML
# =========================

for president, df_p in df_all.groupby("president_short"):

    for speech, df_s in df_p.groupby("speech_label"):
        df_s = df_s.sort_values("block_idx")

        fig = go.Figure()

        for axis in AXES:
            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{axis}"],
                    mode="lines+markers",
                    name=axis,
                    hovertemplate=(
                        "Block: %{x}<br>"
                        f"{axis} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

        fig.add_hline(y=0, line_dash="dash", line_width=1)

        fig.update_layout(
            title=f"{president} - {speech} | PC1-PC5 + GPY Projection Scores",
            xaxis_title="Block index",
            yaxis_title="Projection score",
            template="plotly_white",
            width=1200,
            height=650,
            legend_title="Axis",
            hovermode="x unified"
        )

        out_html = OUT_DIR / f"{safe_filename(president)}_{safe_filename(speech)}_PC1_PC5_GPY_projection.html"
        fig.write_html(out_html, include_plotlyjs="cdn")

print("Konuşma bazlı Plotly HTML dosyaları kaydedildi.")


# =========================
# 7) HER BAŞKAN İÇİN HER EKSENİ 5 KONUŞMADA KARŞILAŞTIRAN HTML
# =========================

for president, df_p in df_all.groupby("president_short"):

    for axis in AXES:
        fig = go.Figure()

        for speech, df_s in df_p.groupby("speech_label"):
            df_s = df_s.sort_values("block_idx")

            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{axis}"],
                    mode="lines+markers",
                    name=str(speech),
                    hovertemplate=(
                        "Block: %{x}<br>"
                        f"{axis} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

        fig.add_hline(y=0, line_dash="dash", line_width=1)

        fig.update_layout(
            title=f"{president} | {axis} Projection Scores Across Speeches",
            xaxis_title="Block index",
            yaxis_title=f"{axis} projection score",
            template="plotly_white",
            width=1200,
            height=650,
            legend_title="Speech",
            hovermode="x unified"
        )

        out_html = OUT_DIR / f"{safe_filename(president)}_{axis}_all_speeches_projection.html"
        fig.write_html(out_html, include_plotlyjs="cdn")

print("Başkan bazlı eksen Plotly HTML dosyaları kaydedildi.")


# =========================
# 8) MASTER HTML: TÜM BAŞKANLAR, TÜM KONUŞMALAR, PC1-PC5 + GPY
# =========================

fig = go.Figure()

for president, df_p in df_all.groupby("president_short"):
    for speech, df_s in df_p.groupby("speech_label"):
        df_s = df_s.sort_values("block_idx")

        for axis in AXES:
            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{axis}"],
                    mode="lines",
                    name=f"{president} | {speech} | {axis}",
                    visible=True,
                    hovertemplate=(
                        f"{president}<br>"
                        f"{speech}<br>"
                        "Block: %{x}<br>"
                        f"{axis} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

fig.add_hline(y=0, line_dash="dash", line_width=1)

fig.update_layout(
    title="All Presidents | All Speeches | PC1-PC5 + GPY Projection Scores",
    xaxis_title="Block index",
    yaxis_title="Projection score",
    template="plotly_white",
    width=1500,
    height=800,
    legend_title="President | Speech | Axis",
    hovermode="x unified"
)

master_html = OUT_DIR / "MASTER_all_presidents_all_speeches_PC1_PC5_GPY_projection.html"
fig.write_html(master_html, include_plotlyjs="cdn")

print("Master Plotly HTML dosyası kaydedildi:")
print(master_html)


# =========================
# 9) KONUŞMA BAZLI ÖZET TABLO
# =========================

score_cols = [f"score_{axis}" for axis in AXES]

summary = (
    df_all
    .groupby(["president_short", "speech_label"])[score_cols]
    .agg(["mean", "median", "std", "min", "max"])
)

summary_csv = OUT_DIR / "PC1_PC5_GPY_projection_summary_by_speech.csv"
summary.to_csv(summary_csv)

print("Özet tablo kaydedildi:")
print(summary_csv)

print("Tüm çıktılar:")
print(OUT_DIR)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pathlib import Path

DATA_FILES = {
    "Fillmore": ".\\selected_speeches\\fillmore_core5_all_blocks_emfd_vectors.csv",
    "Ford": ".\\selected_speeches\\ford_core5_all_blocks_emfd_vectors.csv",
    "Truman": ".\\selected_speeches\\truman_core5_all_blocks_emfd_vectors.csv",
    "Eisenhower": ".\\selected_speeches\\eisenhower_core5_all_blocks_emfd_vectors.csv",
}

PCA_FILE = ".\\pca_export\\pca_export_summary.xlsx"

OUT_DIR = Path(".\\pc1_pc5_projection_plotly")
OUT_DIR.mkdir(exist_ok=True)

FEATURES = ["care_p", "fairness_p", "loyalty_p", "authority_p", "sanctity_p"]
PCS = ["PC1", "PC2", "PC3", "PC4", "PC5"]


# =========================
# PCA LOADINGS VE SCALER
# =========================

loadings_df = pd.read_excel(PCA_FILE, sheet_name="loading_matrix_aligned")
scaler_df = pd.read_excel(PCA_FILE, sheet_name="scaler_stats")

loadings_df = loadings_df.set_index("feature").loc[FEATURES]
scaler_df = scaler_df.set_index("feature").loc[FEATURES]

means = scaler_df["mean"].values.astype(float)
scales = scaler_df["scale"].values.astype(float)

pc_vectors = {
    pc: loadings_df[pc].values.astype(float)
    for pc in PCS
}


# =========================
# PC1-PC5 PROJECTION SCORE HESABI
# =========================

all_dfs = []

for president, path in DATA_FILES.items():
    df = pd.read_csv(path)

    X_raw = df[FEATURES].values.astype(float)
    X_z = (X_raw - means) / scales

    for pc, v in pc_vectors.items():
        df[f"score_{pc}"] = X_z @ v

    df["president_short"] = president
    all_dfs.append(df)

df_all = pd.concat(all_dfs, ignore_index=True)

out_csv = OUT_DIR / "four_presidents_PC1_PC5_projection_scores.csv"
df_all.to_csv(out_csv, index=False)

print("Projection score dosyası kaydedildi:")
print(out_csv)


# =========================
# GÜVENLİ DOSYA ADI
# =========================

def safe_filename(x):
    return (
        str(x)
        .replace("/", "-")
        .replace("\\", "-")
        .replace(":", "-")
        .replace("*", "-")
        .replace("?", "")
        .replace('"', "")
        .replace("<", "")
        .replace(">", "")
        .replace("|", "")
        .replace("\n", " ")
        .strip()
    )


# =========================
# HER KONUŞMA İÇİN PC1-PC5 TEK HTML
# =========================

for president, df_p in df_all.groupby("president_short"):

    for speech, df_s in df_p.groupby("speech_label"):
        df_s = df_s.sort_values("block_idx")

        fig = go.Figure()

        for pc in PCS:
            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{pc}"],
                    mode="lines+markers",
                    name=pc,
                    hovertemplate=(
                        "Block: %{x}<br>"
                        f"{pc} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

        fig.add_hline(y=0, line_dash="dash", line_width=1)

        fig.update_layout(
            title=f"{president} - {speech} | PC1-PC5 Projection Scores",
            xaxis_title="Block index",
            yaxis_title="PCA projection score",
            template="plotly_white",
            width=1200,
            height=650,
            legend_title="PC axis",
            hovermode="x unified"
        )

        out_html = OUT_DIR / f"{safe_filename(president)}_{safe_filename(speech)}_PC1_PC5_projection.html"
        fig.write_html(out_html, include_plotlyjs="cdn")

print("Konuşma bazlı Plotly HTML dosyaları kaydedildi.")


# =========================
# HER BAŞKAN İÇİN HER PC'Yİ 5 KONUŞMADA KARŞILAŞTIR
# =========================

for president, df_p in df_all.groupby("president_short"):

    for pc in PCS:
        fig = go.Figure()

        for speech, df_s in df_p.groupby("speech_label"):
            df_s = df_s.sort_values("block_idx")

            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{pc}"],
                    mode="lines+markers",
                    name=str(speech),
                    hovertemplate=(
                        "Block: %{x}<br>"
                        f"{pc} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

        fig.add_hline(y=0, line_dash="dash", line_width=1)

        fig.update_layout(
            title=f"{president} | {pc} Projection Scores Across Speeches",
            xaxis_title="Block index",
            yaxis_title=f"{pc} projection score",
            template="plotly_white",
            width=1200,
            height=650,
            legend_title="Speech",
            hovermode="x unified"
        )

        out_html = OUT_DIR / f"{safe_filename(president)}_{pc}_all_speeches_projection.html"
        fig.write_html(out_html, include_plotlyjs="cdn")

        out_png = OUT_DIR / f"{safe_filename(president)}_{safe_filename(speech)}_PC1_PC5_projection.png"
        fig.write_image(
            out_png,
            format="png",
            width=2400,
            height=1300,
            scale=2
        )

print("Başkan bazlı PC Plotly HTML dosyaları kaydedildi.")


# =========================
# MASTER HTML
# =========================

fig = go.Figure()

for president, df_p in df_all.groupby("president_short"):
    for speech, df_s in df_p.groupby("speech_label"):
        df_s = df_s.sort_values("block_idx")

        for pc in PCS:
            fig.add_trace(
                go.Scatter(
                    x=df_s["block_idx"],
                    y=df_s[f"score_{pc}"],
                    mode="lines",
                    name=f"{president} | {speech} | {pc}",
                    visible=True,
                    hovertemplate=(
                        f"{president}<br>"
                        f"{speech}<br>"
                        "Block: %{x}<br>"
                        f"{pc} score: " + "%{y:.4f}<extra></extra>"
                    )
                )
            )

fig.add_hline(y=0, line_dash="dash", line_width=1)

fig.update_layout(
    title="All Presidents | All Speeches | PC1-PC5 Projection Scores",
    xaxis_title="Block index",
    yaxis_title="PCA projection score",
    template="plotly_white",
    width=1500,
    height=800,
    legend_title="President | Speech | PC",
    hovermode="x unified"
)

master_html = OUT_DIR / "MASTER_all_presidents_all_speeches_PC1_PC5_projection.html"
fig.write_html(master_html, include_plotlyjs="cdn")

master_png = OUT_DIR / "MASTER_all_presidents_all_speeches_PC1_PC5_projection.png"

fig.write_image(
    master_png,
    format="png",
    width=3000,
    height=1600,
    scale=2
)

print("Master Plotly HTML dosyası kaydedildi:")
print(master_html)


# =========================
# ÖZET TABLO
# =========================

score_cols = [f"score_{pc}" for pc in PCS]

summary = (
    df_all
    .groupby(["president_short", "speech_label"])[score_cols]
    .agg(["mean", "median", "std", "min", "max"])
)

summary_csv = OUT_DIR / "PC1_PC5_projection_summary_by_speech.csv"
summary.to_csv(summary_csv)

print("Özet tablo kaydedildi:")
print(summary_csv)

print("Tüm çıktılar:")
print(OUT_DIR)
```
